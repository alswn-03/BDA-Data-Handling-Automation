```python
from pyspark.sql import SparkSession

## Session 생성
spark = (
    SparkSession
    .builder
    .appName("First Session")
    .getOrCreate()
)

spark
```

```python
spark.sparkContext.getConf().getAll()
```

```python
df = spark.range(500).toDF("number") # DataFrame 생성
```

```python
df.printSchema() # 논리 계획 : 스키마 출력 (아직 물리적 실행 X)
```

```python
## Spark 의 편의 중 하나는, dataframe을 SQL query로 다룰 수 있다는 점이다.
```

```python
## Spark SQL

df.createOrReplaceTempView("df_number") # SQL로 사용하기 위해 View 해준다.

spark.sql("""SELECT * FROM df_number;""").show() # show() = 20개만 출력
```

```python
### Pandas UDF (User Defined Function)
from pyspark.sql.functions import PandasUDFType
from pyspark.sql.functions import pandas_udf
from pyspark.sql.types import IntegerType
from pyspark.sql.functions import col
import pandas as pd


@pandas_udf("short") # 임의의 pandas 코드를 넣을 수 있다.
def double(x):
    return x*2
```

```python
df_with_double = df.withColumn("double_number", double(col("number")))
df_with_double.show()
```
